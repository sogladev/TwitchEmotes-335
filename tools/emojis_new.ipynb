add_dir convert .tga and create entries for a given subfolder

```python
import math
from pathlib import Path
from PIL import Image

def nearest_power_of_2(x):
    return 2 ** int(math.ceil(math.log2(x)))

add_dir = Path('~/tmp/Emojis_Results/Add/').expanduser()
subfolder = "Chromiecraft"

defaultpack_entries = []
emoticons_entries = []

for file in add_dir.iterdir():
    if file.is_file():
        name = file.stem
        img = Image.open(file)
        width, height = img.size

        # Resize to height 32, preserve aspect ratio
        if height != 32:
            new_width = int(width * (32 / height))
            img_resized = img.convert("RGBA").resize((new_width, 32), Image.LANCZOS)
            width, height = img_resized.size
        else:
            img_resized = img.convert("RGBA")

        # Pad to nearest power of 2 for WoW compatibility
        pad_width = nearest_power_of_2(width)
        pad_height = nearest_power_of_2(height)
        padded = Image.new("RGBA", (pad_width, pad_height), (0, 0, 0, 0))
        padded.paste(img_resized, ((pad_width - width) // 2, (pad_height - height) // 2))

        # Save as TGA
        tga_path = add_dir / f"{name}.tga"
        padded.save(tga_path)

        # Get size for LUA string
        defaultpack_entries.append(
            f'    ["{name}"] = "Interface\\\\AddOns\\\\TwitchEmotes\\\\Emotes\\\\{subfolder}\\\\{name}.tga:{pad_height}:{pad_width}",'
        )
        emoticons_entries.append(f'    ["{name}"] = "{name}",')

print("-- TwitchEmotes_defaultpack")
print("TwitchEmotes_defaultpack = {")
print(f"    -- {subfolder}")
for entry in defaultpack_entries:
    print(entry)
print("}")

print("\n-- TwitchEmotes_emoticons")
print("TwitchEmotes_emoticons = {")
for entry in emoticons_entries:
    print(entry)
print("}")
```
